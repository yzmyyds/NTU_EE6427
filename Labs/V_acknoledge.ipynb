```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import clear_output

video_path = "/Users/zimin//VScode_Project/EE6427_Vedio_Signal_Processing/Labs/Videos/Video1.mp4"
cap = cv2.VideoCapture(video_path)

fps = cap.get(cv2.CAP_PROP_FPS)      
frame_count_1s = int(fps * 1)        
frame_idx = 0

while True:
    ret, frame = cap.read()
    if not ret or frame_idx >= frame_count_1s:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame_rgb)
    plt.axis('off')
    clear_output(wait=True)
    plt.pause(1/fps)   
    frame_idx += 1

cap.release()
plt.close()
```
